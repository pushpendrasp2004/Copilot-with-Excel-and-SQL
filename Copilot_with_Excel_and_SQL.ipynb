Question 1: Create table and insert 10,000 rows

- -- Create table
CREATE TABLE SalesData (
    CustomerID INT,
    Name VARCHAR(50),
    Age INT,
    City VARCHAR(50),
    PurchaseAmount DECIMAL(10,2),
    PurchaseDate DATE
);

-- Insert random data (example using loop)
DECLARE @i INT = 1;

WHILE @i <= 10000
BEGIN
    INSERT INTO SalesData
    VALUES (
        @i,
        CONCAT('Customer', @i),
        FLOOR(RAND() * 50) + 18,
        CASE
            WHEN RAND() < 0.2 THEN 'Delhi'
            WHEN RAND() < 0.4 THEN 'Mumbai'
            WHEN RAND() < 0.6 THEN 'Bhopal'
            WHEN RAND() < 0.8 THEN 'Indore'
            ELSE 'Chennai'
        END,
        ROUND(RAND() * 5000, 2),
        DATEADD(DAY, -FLOOR(RAND()*365), GETDATE())
    );

    SET @i = @i + 1;
END;

Question 2: Total sales per city & Top 5 cities

- -- Total sales per city
SELECT City, SUM(PurchaseAmount) AS TotalSales
FROM SalesData
GROUP BY City;

-- Top 5 cities by revenue
SELECT TOP 5 City, SUM(PurchaseAmount) AS TotalRevenue
FROM SalesData
GROUP BY City
ORDER BY TotalRevenue DESC;

Question 3: Customers with above average purchase

- SELECT *
FROM SalesData
WHERE PurchaseAmount > (
    SELECT AVG(PurchaseAmount) FROM SalesData
);

Question 10: VBA Code for Data Cleaning

- Sub CleanData()

    ' Remove duplicates
    ActiveSheet.UsedRange.RemoveDuplicates Columns:=Array(1,2,3,4,5), Header:=xlYes

    ' Fill blank cells
    Dim cell As Range
    For Each cell In ActiveSheet.UsedRange
        If IsEmpty(cell.Value) Then
            cell.Value = "N/A"
        End If
    Next cell

    ' Format table
    ActiveSheet.UsedRange.Columns.AutoFit

    MsgBox "Data Cleaned Successfully!"

End Sub
